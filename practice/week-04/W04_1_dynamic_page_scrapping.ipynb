## 1. 다음 증권 종목 토론실 글 수집

```python
import requests
from bs4 import BeautifulSoup 
import pprint
import json
```

```python
#마우스 우 클릭 -> 검사 -> 상단 탭에서 Network선택

url = "https://finance.daum.net/quotes/A105560#talks"
res = requests.get(url)
res.text
```

```python
#게시판을 요청하는 진짜 URL
url = "https://finance.daum.net/content/debates/A105560?symbolCode=A105560&page=2&perPage=30&notice=true&pagination=true"
res = requests.get(url)
res.text
#403 Error의 의미는 접근 권한이 없다는 뜻입니다.
```

```python
#브라우저에서는 요청이 되므로, 브라우저인 척을 해봅시다!
headers = {
    "referer": "https://finance.daum.net/quotes/A105560",
    "User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36"
}
res = requests.get(url,headers=headers)
data = json.loads(res.text)
pprint.pprint(data)
```

```python
for post in data['data']['posts']:
    print(post['createdAt'], post['title'], post['content'][:100],"...")
```
