```python
import pymysql.cursors
#데이터베이스 접근을 위한 정보를 입력합니다.
HOST = ""
USER = ""
PASSWORD = ""
DATABASE = ""

# Connect to the database
connection = pymysql.connect(host=HOST,
                             port=3306,
                             user=USER,
                             password=PASSWORD,
                             db=DATABASE,
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)

try:
    with connection.cursor() as cursor:
        # Select the records
        # articles라는 테이블에서 조건 없이 10개의 행의 모든 컬럼값.
        sql = "SELECT * FROM `articles` WHERE 1 LIMIT 10"
        cursor.execute(sql)
        result = cursor.fetchall()
        print(result)


    connection.commit()

finally:
    connection.close()
```
